# Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Dataset

```python
data.drop_duplicates(keep='first',inplace=True)
data
```

# Visualization the Dataset

```python
sns.heatmap(corr,annot=True)
```

```python
sns.countplot(data['Ship Mode'])
plt.show()
```

```python
sns.countplot(data['Segment'])
plt.show()
```

```python
data.hist(figsize=(12,12),bins=50)
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.scatter(x=data['Sales'],y=data['Profit'],c='m')
plt.title("Sale VS Profit")
plt.xlabel('Total Sale')
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data['State'],order=(data['State'].value_counts().head(30)).index)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,7))
data['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
# total profit and sales
plt.figure(figsize=(13,9))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
# Counts of Sub-Category Region-Wise
plt.figure(figsize=(12,7))
sns.countplot(x='Sub-Category',hue='Region',data=data)
plt.xticks(rotation=90)
plt.show()
```
